```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clean_file_path ='/content/Frailty_clean.csv'
data = pd.read_csv(clean_file_path, encoding='latin-1')
```

```python
data.head()
```

```python
data = data.select_dtypes(['object']) #strip the trailing spaces
data[data.columns] = data.apply(lambda x: x.str.strip())
data.columns = data.columns.str.strip()
```

```python
data["Frailty"]=data["Frailty"].map({'Y': 1, 'N': 0})
print(data.head())
```

```python
data.dropna(inplace=True)
data["Grip_strength_pounds"] = data["Grip strength"].astype(float) * 2.20
data.to_csv(clean_file_path, index=False)
```

```python
from scipy import stats
t_test_result = stats.ttest_ind(data["Grip_strength_pounds"][data["Frailty"] == 1],
                                data["Grip_strength_pounds"][data["Frailty"] == 0],
                                equal_var=False)
print(t_test_result)
print("T-test statistic:", t_test_result.statistic)
print("P-value:", t_test_result.pvalue)
```
